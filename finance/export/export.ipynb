```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('2019 NIGERIAN IMPORT DATA.xlsx')
df.shape
```

```python
df.head()
```

```python
# let's check unique values
columns=df.columns
for col in columns:
    print(col,':\n',df[col].unique())
    print(df[col].value_counts())
    print('\n',20*'**','\n')
```

```python
df.info()
```

```python
#Check For Duplicates
duplicates = df.duplicated()
duplicates.sum()
```

```python
df.isna().sum()
```

```python
df = df.fillna(df.mode().iloc[0])
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.drop([ 'S/N', 'SCANNING_COY', 'BANK_CODE', 'BANK_NUMBERING', 'MFNUM','FOB_VAL', 'FREIGHT_VAL','USD EQUIV', 'Bank Code', 'Bank name', 'FOB Value Naira', 'CFR'],axis = 1,inplace = True)
```

```python
df.head()
```

#### Export the Clean Data

```python
# Export current clean dataframe
df.to_csv("Clean Nigerian Export df.csv" , index=False)
```
